```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_regression, make_classification
import setuptools.dist
from setuptools.dist import Distribution

#create
scalar = tf.constant(42)  #Скаляр
vector = tf.constant([1.0, 2.0, 3.0])  #Вектор
matrix = tf.constant([[1, 2], [3, 4]])  #Матриця

print("Скаляр:", scalar.numpy())
print("Вектор:", vector.numpy())
print("Матриця:\n", matrix.numpy())

#read
scalar_value = scalar.numpy()
vector_value = vector.numpy()
matrix_value = matrix.numpy()

#update
updated_matrix = tf.add(matrix, 5)
print("Оновлена матриця:\n", updated_matrix.numpy())
updated_scalar = tf.add(scalar, 10)
print("Оновлений скаляр:", updated_scalar.numpy())
updated_vector = tf.add(vector, 2)
print("Оновлений вектор:", updated_vector.numpy())

#delete
del scalar, vector, matrix

#Регресія
X_reg, y_reg = make_regression(n_samples=100, n_features=1, noise=10, random_state=42)
X_reg_tensor = tf.convert_to_tensor(X_reg, dtype=tf.float32)
y_reg_tensor = tf.convert_to_tensor(y_reg, dtype=tf.float32)

#Класифікація
X_clf, y_clf = make_classification(
    n_samples=100, 
    n_features=5,
    n_classes=2, 
    n_clusters_per_class=1, 
    n_informative=3,
    n_redundant=1,
    random_state=42
)
X_clf_tensor = tf.convert_to_tensor(X_clf, dtype=tf.float32)
y_clf_tensor = tf.convert_to_tensor(y_clf, dtype=tf.int32)

#Регресія
plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
plt.scatter(X_reg, y_reg, color='blue', label='Дані регресії')
plt.title("Дані для регресії")
plt.xlabel("X")
plt.ylabel("y")
plt.legend()

#Класифікація
plt.subplot(1, 2, 2)
plt.scatter(X_clf[:, 0], X_clf[:, 1], c=y_clf, cmap='coolwarm', label='Дані класифікації')
plt.title("Дані для класифікації")
plt.xlabel("Ознака 1")
plt.ylabel("Ознака 2")
plt.legend()
plt.tight_layout()
plt.show()
```
